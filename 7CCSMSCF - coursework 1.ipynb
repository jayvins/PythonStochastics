## Exercise 18

Follwing David Goldberg's paper, write a Python function that implements the quadratic formula for finding the roots of a quadratic function, in such a way that it does not suffer from at least one of the two potential sources of catastrophic cancellation.



```python
def quad_solve( f ):
    #f is a quadratic function 
    
    #use matrix M to extract coefficients from f, and coerce into array
    M = np.matrix([[0,0,1],[1,1,1],[4,2,1]])    
    coef_vect = np.array( np.dot( M.I, [f(0),f(1),f(2)]) )[0]
    
    ret_vec = list()
    
    
    #calculate the discriminant to see if there exist real solutions
    disc = coef_vect[1]**2 - 4*coef_vect[0]*coef_vect[2]
   
    if(disc < 0):
        return print("complex solutions")
    elif( coef_vect[1] < 0 ):
        ret_vec.append((-coef_vect[1]+(disc ** 0.5))/(2*coef_vect[0]))
        ret_vec.append((2*coef_vect[2])/(-coef_vect[1]+(disc ** 0.5)))
    else:
        ret_vec.append((2*coef_vect[2])/(-coef_vect[1]-(disc ** 0.5)))
        ret_vec.append((-coef_vect[1]-(disc ** 0.5))/(2*coef_vect[0]))
    return np.array(ret_vec)          
        
```
